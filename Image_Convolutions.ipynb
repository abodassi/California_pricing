## Blur using Kernals

Kernals are the subset of the image matrix. They can be of MxN size where M and N are odd intergers. We will use a 3x3 kernal using NumPy

To blur an image, a 3x3 matrix of 1s is used and divided by a factor. Factor is an integer that specifies the amount of blurring. We will use 15 as the factor.

filter2D() function can be used to apply Kernal to an image

OpenCV has provided blur() method to blur an image

## Gaussian Blur

In this type, the image is blurred and the contrast it reduced


## Median Blur

It is similar to blurring using Kernal, the central element of an image is replaced by median of all pixels

## Sharpen using Kernel

We replace the central element of a Image by an amplified version of the same.

## Bilateral Filtering
It is highly effective in noise removal while keeping edges sharp.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/download.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

kernel = np.ones((5, 5), np.float32) / 25
kernel_blur = cv2.filter2D(image, -1, kernel)

avg_blur = cv2.blur(image, (5, 5))
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)
median_blur = cv2.medianBlur(image, 5)

sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])
sharpen = cv2.filter2D(image, -1, sharpen_kernel)

bilateral = cv2.bilateralFilter(image, 9, 75, 75)

plt.figure(figsize=(15, 10))

titles = ['Original', 'Kernel Blur', 'Average Blur', 'Gaussian Blur', 'Median Blur', 'Sharpen', 'Bilateral Filter']
images = [image, kernel_blur, avg_blur, gaussian_blur, median_blur, sharpen, bilateral]

for i in range(len(images)):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
